```python
#creating primary_sector column
temp=master_Frame["category_list"]
master_Frame["primary_sector"]=temp.str.split("|",expand=True)[0]#breaking category list from | and storing first value
master_Frame
```

```python
Mapping=pd.read_csv('mapping.csv')# reading mapping.csv
Mapping
```

```python
Mapping = pd.melt(Mapping,id_vars=["category_list"],var_name='Main_Sector',value_name = "TrueFalse")
Mapping = Mapping[Mapping.TrueFalse != 0]#saving for value 1
```

```python
# cleaning data by checking spellings like 0lytics to analytics
Mapping['category_list']=Mapping['category_list'].astype(str).apply(lambda x: x if x=="Enterprise 2.0" else x.replace('0', 'na')).str.lower()#cleanind strings in category list by replacing 0 in sting with na accorging to pattern except in enterprise 2.0 and changing into lower case
Mapping
```

```python
master_Frame["primary_sector"]=master_Frame["primary_sector"].str.lower()#changing master_Frame category_list column in lower case 
master_Frame
```

```python
#merging on primary sector
master_Frame = pd.merge(master_Frame,Mapping.filter(['category_list','Main_Sector'],axis=1),left_on="primary_sector",right_on="category_list")#merging mapping frame into master_Frame
master_Frame
#master_count=master_Frame.category_list.value_counts().count()
#master_count
```

```python
master_Frame=master_Frame.drop(['category_list_y'],axis=1)# dropping dupicated column
master_Frame=master_Frame.rename(columns={'category_list_x': 'category_list'})# renaming the column
master_Frame
```

```python

master_Frame=master_Frame[master_Frame.Main_Sector!='Blanks']#blank is not any sector therfore dropping it
```

```python
# DF1 of top first country "USA" of venture type found "venture" with usd amt in between 5 to 12 million usd(inclusive)
DF1=master_Frame[(master_Frame.country_code=="USA")&(master_Frame.funding_round_type=="venture")&(master_Frame.raised_amount_usd>=5000000)&(master_Frame.raised_amount_usd<=15000000)]
DF1
```

```python
# DF2 of top second country "GBR" of venture type found "venture" with usd amt in between 5 to 12 million usd(inclusive)
DF2=master_Frame[(master_Frame.country_code=="GBR")&(master_Frame.funding_round_type=="venture")&(master_Frame.raised_amount_usd>=5000000)&(master_Frame.raised_amount_usd<=15000000)]
DF2
```

```python
# DF3 of top third country "IND" of venture type found "venture" with usd amt in between 5 to 12 million usd(inclusive)
DF3=master_Frame[(master_Frame.country_code=="IND")&(master_Frame.funding_round_type=="venture")&(master_Frame.raised_amount_usd>=5000000)&(master_Frame.raised_amount_usd<=15000000)]
DF3
```

```python
# adding column "Total_no_investmnt" with value count of raised_amt_usd for DF1
total_no_investmnt=DF1.groupby("Main_Sector")["raised_amount_usd"].count().rename("Total_no_investmnt").reset_index()
DF1=DF1.merge(total_no_investmnt)
DF1
```

```python
# adding column "Total_amt_invested" with value sum of raised_Amt_usd for DF1
total_amt_invested=DF1.groupby("Main_Sector")["raised_amount_usd"].sum().rename("Total_amt_invested").reset_index()
DF1 = DF1.merge(total_amt_invested)
DF1["Total_amt_invested"]=DF1["Total_amt_invested"].apply(lambda x:x/1000000)
DF1
```

```python
DF1.to_csv('USA.csv',sep=',',encoding='utf-8',index=False)
```

```python
# adding column "Total_no_investmnt" with value count of raised_Amt_usd for DF2
total_no_investmnt=DF2.groupby("Main_Sector")["raised_amount_usd"].count().rename("Total_no_investmnt").reset_index()
DF2=DF2.merge(total_no_investmnt)
DF2
```

```python
#adding column "Total_amt_invested" with value sum of raised_Amt_usd for DF2
total_amt_invested=DF2.groupby("Main_Sector")["raised_amount_usd"].sum().rename("Total_amt_invested").reset_index()
DF2 = DF2.merge(total_amt_invested)
DF2["Total_amt_invested"]=DF2["Total_amt_invested"].apply(lambda x:x/1000000)
DF2
```

```python
DF2.to_csv('GBR.csv',sep=',',encoding='utf-8',index=False)
```

```python
#adding column "Total_no_investmnt" with value count of raised_Amt_usd for DF3
total_no_investmnt=DF3.groupby("Main_Sector")["raised_amount_usd"].count().rename("Total_no_investmnt").reset_index()
DF3=DF3.merge(total_no_investmnt)
DF3
```
